```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
from mpl_toolkits import mplot3d
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
np.random.seed(42)
x1 = 2 * np.random.rand(200, 1)
x2 = 2 * np.random.rand(200, 1)
ys = 4 + 7 * x1 + 5 * x2 + np.random.rand(200, 1)
```

```
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, ys, cmap='viridis', edgecolor='red', alpha=0.4)
plt.show()
```

```
# split the data into training and test sets
# train set
train_x1 = x1[:160]
train_x2 = x2[:160]
train_ys = ys[:160]

# test set
test_x1 = x1[160:]
test_x2 = x2[160:]
test_ys = ys[160:]
```

**Linear regression using numpy**

```
# number of epochs
epochs = 1000
# learning rate
lr = 0.01
```

```
# initial value for weights w1, w2 and bias
w1 = np.random.randn(1)
w2 = np.random.randn(1)

b = np.zeros(1)
```

```
for epoch in range(epochs):
  for i in range(160):
    y_pred = w1 * train_x1[i] + w2 * train_x2[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_x1[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_x1[i] + w2 * test_x2[i] + b - test_ys[i]) ** 2
test_loss /= 20
```

```
test_loss
```

```
pred_ys = w1 * test_x1 + w2 * test_x2 + b
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(test_x2, test_x2, test_ys, color='blue', alpha=0.6)
ax.scatter(test_x1, test_x2, pred_ys, color='black', alpha=0.6)
ax.set_title('Predictions against Actuals')
plt.show()
```

```
plt.plot(test_x1, test_ys, "b.")
plt.plot(test_x1, pred_ys, "r.") 
plt.xlabel("$x_1$", fontsize=20)
plt.ylabel("$y$", rotation=0, fontsize=20)
plt.axis([0, 2, 0, 30])
plt.show()
```

```
# View from X2 plane
plt.plot(test_x2, test_ys, "b.")
plt.plot(test_x2, pred_ys, "r.") # predicted values
plt.xlabel("$x_2$", fontsize=20)
plt.ylabel("$y$", rotation=0, fontsize=20)
plt.axis([0, 2, 0, 30])
plt.show()
```
